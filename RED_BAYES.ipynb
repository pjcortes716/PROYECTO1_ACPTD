```python
import pandas as pd
```

```python
heart_disease=pd.read_csv("processed.cleveland.data", names=["E","S","D","H","C","A",
                                                            "RE","HR","AI","SER",
                                                            "P","F","T",
                                                            "EC"],na_values='?')
heart_disease
```

```python
heart_disease=heart_disease.dropna()
heart_disease#eliminamos los datos con valores nulos
```

Las claves son:
1. E: EDAD; Edad del paciente en años
2. S: SEXO; 1 hombre, 0 mujer
3. D: DOLOR_PECHO; Tipo de dolor en el pecho; 1 Angina tipica, 2 Angina no tipica, 3, dolor no anginal, 4 asintomatico
4. H: PRESION_SANG; Presion sanguinea (mm HG al llegar al hospital)
5. C: COLESTEROL; serum colesterol en mg/dl
6. A: AZUCAR_SANG; 1 si >120mg/dl, 0 d.l.c
7. RE: RES_ELECTC; 0 si normal, 1 si existe una anormalidad en la onda, 2 showing probable or definite left ventricular hypertrophy by Estes' criteria
8. HR: MAX_HERT_R;
9. AI: EXERCISE_IND_ANG; 1 si se induce, 0 si no
10. SER: ST_EXERC_REPOSO; ST depression induced by exercise relative to rest
11. P: PENDIENTE; slope: the slope of the peak exercise ST segment, 1 upslopging, 2 flat, 3 downsloping
12. F: NUMBER_MV_F; number of major vessels (0-3) colored by flourosopy
13. T: thal; 3 = normal; 6 = fixed defect; 7 = reversable defect
14. EC: num; diagnosis of heart disease (angiographic disease status)
        -- Value 0: < 50% diameter narrowing
        -- Value 1: > 50% diameter narrowing
        (in any major vessel: attributes 59 through 68 are vessels)

```python
#transformamos la columna de D, dolor en el pecho, en binaria
heart_disease.D=heart_disease.D.apply(lambda num: 0 if num==4 else 1)
heart_disease
```

```python
#transformamos ahora la columna de presion sanguinea, tomando como limite un valor de 130 mm de Hg
heart_disease.H=heart_disease.H.apply(lambda num:0 if num<130 else 1)
heart_disease
```

```python
#transformamos ahora la columna C, colesterol, teniendo un valor normal de 200 mg/dl
heart_disease.C=heart_disease.C.apply(lambda num: 0 if num<=200 else 1)
heart_disease
```

```python
#Transformamos la columna RE, resultados del electrocardiograma:
heart_disease.RE=heart_disease.RE.apply(lambda num: 0 if num==0 else 1)
heart_disease
```

```python
#tranformamos ahora la columna HR, maxima frecuencia cardiaca, tomando como valor minimo normal 150
heart_disease.HR=heart_disease.HR.apply(lambda num: 0 if num>150 else 1)
heart_disease
```

```python
#Ahora transforamos la columna SER, depresion inducida por la prueba de esfuerzo, si es mayor a 1 mm se considera como psitivo
heart_disease.SER=heart_disease.SER.apply(lambda num: 0 if num<=1 else 1)
heart_disease
```

```python
#Modificamos ahora la columna de P, correpondiente a la pendiente de pico de la curva de la prueba de esfuerzo: Si es creciente
#tomara un valor de cero, de lo contrario 1

heart_disease.P=heart_disease.P.apply(lambda num: 0 if num==1 else 1)
heart_disease
```

```python
#modificamos la columna F: si se presenta algun vaso coloreado 1, de lo contrario cero

heart_disease.F=heart_disease.F.apply(lambda num: 0 if num==0 else 1)
heart_disease
```

```python
#modificamos ahora la columna T: si presenta cualquier grado de talasemia 1, de lo contrario 0:
heart_disease["T"]=heart_disease["T"].apply(lambda num: 0 if num==3 else 1)
heart_disease
```

```python
#transformamos ahora la la columna EC:
heart_disease.EC=heart_disease.EC.apply(lambda number: 0 if number==0 else 1)
heart_disease
```

```python
#fijamos una edad de bajo riesgo para sufrir enfermedades cardiacas por debajo de 60 años:
heart_disease.E=heart_disease.E.apply(lambda number: 0 if number<=60 else 1)
heart_disease
```
